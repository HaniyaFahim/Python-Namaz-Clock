```python
import tkinter as tk
from tkinter import *
import time

background_color = input("Enter Background Colour: ")
text_color = input("Enter Text Colour: ")

root = Tk()
root.geometry('800x800')
root.title('Namaz Clock')
root.configure(bg='lightblue')


def tick(time1=""):
    time2 = time.strftime('%I : %M : %S')
    clock_frame.configure(text=time2)
    clock_frame.after(20,tick)
    
def fajar_namaz():
    fajar.configure(text="04:45 am Fajar")

def zuhar_namaz():
    zuhar.configure(text="01:00 pm Zuhar")

def asar_namaz():
    asar.configure(text="05:20 pm Asar")

def magrib_namaz():
    magrib.configure(text="07:35 pm Magrib")
    
def isha_namaz():
    isha.configure(text="08:27 pm Isha")
    
def jumma_namaz():
    jumma.configure(text="02:00 pm Jumma")
    
    
    


clock_frame = Label(root, font=(  'Segoe Print',90,'bold'),bg=background_color,fg=text_color , borderwidth=20, relief="ridge",padx=-10,pady=-10 )
clock_frame.pack(fill='both',expand=1)

fajar = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color,  borderwidth=2, relief="ridge",padx=10,pady=10, anchor='n')
fajar.pack(fill=X)

zuhar = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color, borderwidth=2, relief="ridge",padx=10,pady=10 , anchor='n')
zuhar.pack(fill=X)

asar = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color, borderwidth=2, relief="ridge",padx=10,pady=10 , anchor='n')
asar.pack(fill=X)

magrib = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color, borderwidth=2, relief="ridge" ,padx=10,pady=10, anchor='n')
magrib.pack(fill=X)

isha = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color, borderwidth=2, relief="ridge",padx=10,pady=10 , anchor='n')
isha.pack(fill=X)

jumma = Label(root, font=("Arial",20,'bold'),bg=background_color,fg=text_color, borderwidth=2, relief="ridge",padx=10,pady=10 , anchor='n')
jumma.pack(fill=X)

tick()
fajar_namaz()
zuhar_namaz()
asar_namaz()
magrib_namaz()
isha_namaz()
jumma_namaz()




root.mainloop()bg=background_color,fg=text_color, borderwidth=2, relief="ridge",padx=10,pady=10 
```
